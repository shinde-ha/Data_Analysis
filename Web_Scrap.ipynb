```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://www.primark.com/en-us/products/category/men'
```

```python
#checks whether link is valid(200) or not
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
bool_valid = bool(response.status)
bool_valid
```

```python

soup=BeautifulSoup(response.data, "lxml")
```

```python
soup
```

```python

items = soup.find_all('a')
items
```

```python
for item_attribute in items:
    names = item_attribute.contents[:]
    links = item_attribute.get('href')
    print(names)
    print(links)
```

```python
boolValid = bool(links)
boolValid
```

```python
items[0]
```

```python
import datetime
print ("Current date and time: " , datetime.datetime.now().strftime("%b %d %Y %H:%M:%S"))
```

```python

links=[]
#try:
for url2 in items:
    tmp=[]
    tmp.append(url2.get('href'))
    tmp.append(url2.text)
    tmp.append(boolValid)
    tmp.append(datetime.datetime.now().strftime("%b %d %Y %H:%M:%S"))
    links.append(tmp)
    print(tmp)
```

```python
links
```
